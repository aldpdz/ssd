```python
def fix_outliers(data):
    new_data = []
    max_value = data[0]
    
    for index, value in enumerate(data):
        if value > max_value:
            new_value = (data[index -1] + data[index + 1]) / 2
            new_data.append(new_value)
        else:
            new_data.append(value)
    return new_data
```
