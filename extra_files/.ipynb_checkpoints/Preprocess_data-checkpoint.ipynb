```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import json
import helper
from skimage import io
from sklearn.model_selection import train_test_split
```

```python
path_root = '/home/aldo/Documents/data-cic/'
path_to_images = path_root + 'images'
path_process_data = path_root + 'preprocess_data'
path_to_train = path_process_data + '/img_train'
path_to_test = path_process_data + '/img_test'
path_to_val = path_process_data + '/img_val'
```

```python
# Load json file with ground truth
with open(path_root + 'via_region_data _final.json', 'r') as json_file:
    dict_json = json.load(json_file)
new_json = helper.clean_dict(dict_json)
```

```python
name_images = list(new_json.keys())
```

```python
# Split data into training and test
name_images_train, name_img_eval, = train_test_split(name_images, test_size=.3)
```

```python
# Split data into test and validation
name_images_test = name_img_eval[:45]
name_images_val = name_img_eval[45:]
print(len(name_images_test))
print(len(name_images_val))
print(len(name_images_train))

# Save images into the paths
helper.save_images_to(path_to_images, path_to_train, name_images_train)
helper.save_images_to(path_to_images, path_to_test, name_images_test)
helper.save_images_to(path_to_images, path_to_val, name_images_val)
```

```python
# Create dictionaries to save as numpy files
dict_label_train = helper.create_sub_dict(new_json, name_images_train) 
dict_label_test = helper.create_sub_dict(new_json, name_images_test)
dict_label_val = helper.create_sub_dict(new_json, name_images_val)
```

```python
images, labels = helper.load_images_labels(path_to_images, new_json)
images_train, label_train = helper.load_images_labels(path_to_train, dict_label_train)
images_test, label_test = helper.load_images_labels(path_to_test, dict_label_test)
images_val, label_val = helper.load_images_labels(path_to_val, dict_label_val)
```

```python
# Save labels train, test, original
np.save(path_process_data + '/label_train', label_train)
np.save(path_process_data + '/label_test', label_test)
np.save(path_process_data + '/label_val', label_val)
np.save(path_process_data + '/label_all', labels)
```

```python
# Save images 300, 512, original
images_train300 = helper.resize_images(images_train, 300, 300)
images_test300 = helper.resize_images(images_test, 300, 300)
images_val300 = helper.resize_images(images_val, 300, 300)
images_all300 = helper.resize_images(images, 300, 300)
np.save(path_process_data  + '/images_train_300x300', images_train300)
np.save(path_process_data  + '/images_test_300x300', images_test300)
np.save(path_process_data  + '/images_val_300x300', images_val300)
np.save(path_process_data  + '/images_all_300x300', images_all300)

images_train512 = helper.resize_images(images_train, 512, 512)
images_test512 = helper.resize_images(images_test, 512, 512)
images_val512 = helper.resize_images(images_val, 512, 512)
images_all512 = helper.resize_images(images, 512, 512)
np.save(path_process_data  + '/images_train_512x512', images_train512)
np.save(path_process_data  + '/images_test_512x512', images_test512)
np.save(path_process_data  + '/images_val_512x512', images_val512)
np.save(path_process_data  + '/images_all_512x512', images_all512)
```

# Create validation and test for 320x320 images

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import json
import helper
import os
from skimage import io
from sklearn.model_selection import train_test_split
```

```python
path_root = '/home/aldo/Documents/data-cic/'
path_to_images = path_root + 'images'
path_process_data = path_root + 'preprocess_data'
path_to_test = path_process_data + '/img_test'
path_to_val = path_process_data + '/img_val'
```

```python
# Load json file with ground truth
with open(path_root + 'via_region_data _final.json', 'r') as json_file:
    dict_json = json.load(json_file)
new_json = helper.clean_dict(dict_json)
```

```python
# get images from directory
name_images_test = os.listdir(path_to_test)
name_images_val = os.listdir(path_to_val)
```

```python
# Create dictionaries to save as numpy files
dict_label_test = helper.create_sub_dict(new_json, name_images_test)
dict_label_val = helper.create_sub_dict(new_json, name_images_val)
```

```python
# Load images and labels
images_test, label_test = helper.load_images_labels(path_to_test, dict_label_test)
images_val, label_val = helper.load_images_labels(path_to_val, dict_label_val)
```

```python
# Save labels train, test, original
np.save(path_process_data + '/label_test_320', label_test)
np.save(path_process_data + '/label_val_320', label_val)
```

```python
# Save images 320
images_test300 = helper.resize_images(images_test, 320, 320)
images_val300 = helper.resize_images(images_val, 320, 320)
np.save(path_process_data  + '/images_test_320x320', images_test300)
np.save(path_process_data  + '/images_val_320x320', images_val300)
```
