```python
import seaborn as sns
import numpy as np
import helper as hp
import pandas as pd
import matplotlib.pyplot as plt
import os.path
import json
```

```python
path_data = '../../data-cic/'
absolute_path = os.path.abspath(os.path.dirname('__file__'))
path_data = os.path.join(absolute_path, path_data)
```

```python
# Read json that contains bounding boxes from each image
with open(path_data + 'via_region_data _final.json', 'r') as json_file:
    dict_json = json.load(json_file)
clean_json = hp.clean_dict(dict_json)
```

```python
# List of names for training, validation and test
train_list = os.listdir(path_data + '/preprocess_data/img_train')
val_list = os.listdir(path_data + '/preprocess_data/img_val')
test_list = os.listdir(path_data + '/preprocess_data/img_test')
```

```python
# Get all bounding boxes

bounding_boxes = []
bb_validation = []
bb_test = []
bb_training = []

# Get images' names
keys = list(clean_json.keys())
# Iterate over all images' name
for name_img in keys:
    # Get bb from the image
    bb_list = hp.get_ground_truth(clean_json[name_img])
    
    for bb in bb_list:
        # Calculate area of bounding box
        bb.append(bb[0] + (bb[2] // 2)) # center coordinate x
        bb.append(bb[1] + (bb[3] // 2)) # center coordinate y
        bb.append((bb[2] + 1) * (bb[3] + 1)) # area of bounding box
        bb.append(bb[2] / bb[3])
        bounding_boxes.append(bb)
        
        
        if name_img in train_list:
            bb_training.append(bb)
        elif name_img in val_list:
            bb_validation.append(bb)
        elif name_img in test_list:
            bb_test.append(bb)
```

```python
# Number of bounding boxes in training, validation and test
print('Training:', len(bb_training))
print('Validation:', len(bb_validation))
print('Test:', len(bb_test))
print('Total:', (len(bb_training) + len(bb_validation) + len(bb_test)))
```

```python
bounding_boxes = np.array(bounding_boxes)
columns = ['Coordinate x', 'Coordinate y', 'Ancho', 'Alto', 'Center x', 'Center y', 'Area', 'Relación de aspecto']
```

```python
# Create DataFrame for seabron
data = pd.DataFrame(bounding_boxes, columns=columns)
print(data.shape)
```
